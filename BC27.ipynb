```python
# Batting - Standard scrape
time.sleep(5)
content = driver.page_source
soup = BeautifulSoup(content, 'html.parser')
table = soup.find('table',{'class':'gcSortTable'})
headers = []
for i in table.findAll('th'):
    title = i.text
    if '\n' in title:
        title_splits = title.split('\n')
        keep_title = title_splits[0]
        headers.append(keep_title)
    else:
        headers.append(title)
rows = []
for j in table.findAll('tr')[1:]:
    row = []
    for k in j.findAll('td'):
        datapoint = k.text
        if '\n' in datapoint:
            data_splits_n = datapoint.split('\n')
            keep_split_n = data_splits_n[1]
            keep_data_n = keep_split_n.strip()
            if '%' in keep_data_n:
                data_splits_perc = keep_data_n.split('%')
                keep_split_perc = data_splits_perc[0]
                keep_data_perc = float(keep_split_perc)/100.0
                row.append(keep_data_perc)
            else:
                row.append(keep_data_n)
        else:
            row.append(datapoint)
    rows.append(row)
batting_s = pd.DataFrame(rows,columns=headers)
batting_s = batting_s.iloc[:-3,:]
```
